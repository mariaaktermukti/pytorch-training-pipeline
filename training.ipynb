## Design a neural network using nn.Module.

## Write Answer 02:


```python
import torch
import torch.nn as nn
```

```python
class CyberBullyingNN(nn.Module):
    def __init__(self):
        super(CyberBullyingNN, self).__init__()

        # Input layer
        self.fc1 = nn.Linear(4, 8)   # hidden
        self.fc2 = nn.Linear(8, 1)   # output

        # Activationfunction
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)   # hidden activation
        x = self.fc2(x)
        x = self.sigmoid(x)  # binary output
        return x
```

```python
#model creation
model = CyberBullyingNN()
print(model)
```

```python
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print("Total parameters:", total_params)
```

Input features: 4

Hidden layer: 8 neurons

Output layer: 1 neuron


Activation Functions:

Hidden layer: ReLU cz non-linearity, fast learning
Output layer: Sigmoid cz it gives probability between 0 and 1

Parameter Calculation:

Layer 1 :

W: 4 * 8 = 32

B: 8

Total = 32 + 8 = 40

----

Layer 2:

W: 8 * 1 = 8

B: 1

Total = 8 + 1 = 9

-----

Total Trainable Parameters =  40 + 9 = 49 parameters

## Implement a full training loop:

Forward pass

Loss computation

Backward pass

Parameter update

Gradient reset

### Requirements:

Train for at least 100 epochs.

Print loss every 10 epochs.

Store training loss history(You can pick your own Data Structure).

Explain clearly what happens in each step of the pipeline.

## Write Answer 04:

```python
epochs = 100
loss_history = []

for epoch in range(epochs):

    # Forward Pass
    y_pred = model(X_train)

    # Loss Computation
    current_loss = loss(y_pred, y_train)

    # backward Pass
    current_loss.backward()

    # Parameter Update
    optimizer.step()

    # Gradient reset
    optimizer.zero_grad()

    # Store loss
    loss_history.append(current_loss.item())

    # Print every 10 epochs
    if epoch % 10 == 0:
        print("Epoch:", epoch, "Loss:", current_loss.item())
```

Loss is decreasing steadily, so model is learning properly, stable convergence

1. Forward Pass:

We give input data to the model, and it passes through the network to produce a prediction.


2. Loss Computation:

This tells us how wrong the model is.

3. Backward Pass:

This helps the model understand how to improve.

4. Parameter Update:

Because of updation the model learns and try to be more better.

5. Gradient Reset:
It keeps learning clean and correct.

6. Loss Storage:

We save the loss from each epoch in a list.


7. Printing Loss:

Every 10 epochs, we print the loss to check progress.
It is what how we can track if the model is actually learning.

## Write Answer 07:

Why must gradients be reset every epoch?

Because gradients accumulate in PyTorch.
If not reset, old gradients will mix with new ones and give wrong updates.

---

What happens if learning rate is too high?

Model updates become too big.
Loss can be unstable, and can't give proper learning rate at all.

-----

What happens if learning rate is too small?

Model will learns very slowly.
Takes too many epochs to converge. and will be more late for a better performnce.

---

Why define layers in __init__ not in forward()?

__init__ creates layers once.
forward() only uses them for computation.
If defined in forward(), layers reset every time -> as a result, model can't learn properly.
